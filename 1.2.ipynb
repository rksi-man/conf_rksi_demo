# Задание №1
Напишите программу, которая просит пользователя что-нибудь ввести с клавиатуры. Если он вводит какие-
нибудь данные, то на экране должно выводиться сообщение "**ОК**". Если он не вводит данные, а просто
нажимает Enter, то программа ничего не выводит на экран.




```python
# Код начинается здесь ...
user_pass = input('Введите что-то: ')

if len(user_pass) > 0:
  print('OK')
```

# Задание №2
Напишите программу, которая запрашивает у пользователя число. <br>Если оно больше нуля, то в ответ на
экран выводится "positive". <br>Если введенное число не является положительным, то на экран должно выводиться - "negative".





```python
# Код начинается здесь ...
num = int(input('Введите число: '))

if num > 0:
  print('positive')
else:
  print('negative')
```

# Задание №3
Напишите программу, которая запрашивает на ввод число. <br>Если оно положительное, то на экран выводится
цифра 1. <br>Если число отрицательное, выводится -1. <br>Если введенное число – это 0, то на экран выводится 0.<br>
Используйте в коде условный оператор множественного ветвления.





```python
# Код начинается здесь ...
num = int(input('Введите число: '))
# 0
if num > 0:
  print(1)
else:
  if num < 0:
    print(-1)
  else:
    print(0)
```

```python
num = int(input('Введите число: '))
# 10
if num > 0:
  print(1)
elif num < 0:
  print(-1)
else:
  print(0)
```
