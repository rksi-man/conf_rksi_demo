# Задание №2
У Вас имеется некий массив данных:
```
a = [-1, 3, 9, -6, 12, 3, 0, -18, 4]
```
Необходимо создать файл "result_m_24_2.txt" и записать в него массив так, чтобы каждое значение хранилось на отдельной строке





````python
# Код начинается здесь ...
a = [-1, 3, 9, -6, 12, 3, 0, -18, 4]

with open('result_m_24_2.txt', 'w') as fw:
  for i in a:
    fw.write(str(i) + '\n')
````

# Задание №3
У Вас имеется некий массив данных:
```
a = [-1, 3, 9, -6, 12, 3, 0, -18, 4, 'o', 'O', 0, 12, 'e', 'u']
```
Вам необходимо найти в нём отрицательные числа и записать в файл "result_m_24_3.txt" в одну строку разделяя знаком пробел





````python
# Код начинается здесь ...
a = [-1, 3, 9, -6, 12, 3, 0, -18, 4, 'o', 'O', 0, 12, 'e', 'u']

with open('result_m_24_3.txt', 'w') as fw:
  for i in a:
    if isinstance(i, int) and i < 0:
      fw.write(str(i) + ' ')
````

# Задание №4
Создайте файл nums.txt
Заполните файл двадцатью случайными числами в диапазоне от -100 до 100, записанных через пробел.
Напишите программу, которая подсчитывает и выводит на экран общую сумму чисел, хранящихся в этом файле, также создает файл "negative.txt", который должен содержать только отрицательные значения, если таковые имеются. В случае если отрицательные значения будут записаны в файл вывести их в виде списка после собщения: "Negative values were found, reading from file: "


````python
# Код начинается здесь ...
import random


with open('nums.txt', 'w') as fw:
  for _ in range(20):
    fw.write(str(random.randint(-100,100)) + ' ')
````

````python
with open('nums.txt', 'r') as fr:
  list_data = fr.read()
  list_data = list_data.split(' ')
````

````python
for i,v in enumerate(list_data):
  if v:
    list_data[i] = int(v)
  else:
    list_data.pop(i)
````

````python
sum(list_data)
````

````python
list_neg = []

for i in list_data:
  if i < 0:
    list_neg.append(i)
````

````python
if len(list_neg) > 0:
  with open('negative.txt', 'w') as fw:
    for i in list_neg:
      fw.write(str(i) + ' ')
  print(f'Negative values were found, reading from file: {list_neg}')
````
