–î–æ–ø–∏—à–∏—Ç–µ –º–µ—Ç–æ–¥ __add__ –≤ –∫–ª–∞—Å—Å–µ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –æ–ø–µ—Ä–∞—Ü–∏–∏ —Å–ª–æ–∂–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–æ–≤





````python
class Vector:
   def __init__(self, a, b):
      self.a = a
      self.b = b
   def __str__(self):
      return f'Vector ({self.a}, {self.b})'
   # –ö–æ–¥ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –∑–¥–µ—Å—å ...
   def __add__(self, other):
      return str(Vector(self.a + other.a, self.b + other.b))

v1 = Vector(5,12)
v2 = Vector(2,-15)
print(v1 + v2)
````

````python
v1 + v2
````

# –ó–∞–¥–∞–Ω–∏–µ ‚Ññ3
–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Coffee (–¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–∏–¥–∞ –∫–æ—Ñ–µ), <br>–ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π 1 –∞—Ä–≥—É–º–µ–Ω—Ç –ø—Ä–∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ (–æ—Ç–≤–µ—á–∞—é—â–∏–π –∑–∞ –¥–æ–ø–æ–ª–Ω–µ–Ω–∏—è –∫ –≤—ã–±—Ä–∞–Ω–Ω–æ–º—É –≤–∏–¥—É –∫–æ—Ñ–µ). <br>
–í —ç—Ç–æ–º –∫–ª–∞—Å—Å–µ —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –º–µ—Ç–æ–¥ check_my_drink(), –≤—ã–≤–æ–¥—è—â–∏–π –Ω–∞ –ø–µ—á–∞—Ç—å ¬´–ö–æ—Ñ–µ –∏ {–î–û–ü–û–õ–ù–ï–ù–ò–ï}¬ª –≤ —Å–ª—É—á–∞–µ –Ω–∞–ª–∏—á–∏—è –¥–æ–±–∞–≤–∫–∏, <br>–∞ –∏–Ω–∞—á–µ –æ—Ç–æ–±—Ä–∞–∑–∏—Ç—Å—è —Å–ª–µ–¥—É—é—â–∞—è —Ñ—Ä–∞–∑–∞: ¬´–û–±—ã—á–Ω—ã–π –∫–æ—Ñ–µ¬ª.





````python
# –ö–æ–¥ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –∑–¥–µ—Å—å ...
class Coffee:
  def __init__(self, dop = ''):
    self.dop = dop
  def check_my_drink(self):
    if self.dop:
      print(f'–ö–æ—Ñ–µ –∏ {self.dop}')
    else:
      print('–û–±—ã—á–Ω—ã–π –∫–æ—Ñ–µ')
````

````python
#–¢–µ—Å—Ç—ã
drink1 = Coffee()
drink2 = Coffee('–∫–æ–∫–æ—Å–æ–≤—ã–π —Å–∏—Ä–æ–ø')
drink1.check_my_drink()
drink2.check_my_drink()
````
