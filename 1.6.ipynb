# Задание №1
Напишите генератор списка, который заполняет список данными, которые вводит пользователь. Количество вводнимых значений 5. Название переменной генератора: "gen_1" Другой генератор списка должен преобразовывать данные в числа. Если какой-либо элемент первого списка нельзя
преобразовать в число, то во второй список ("gen_2") этот элемент попадать не должен. <br>
Выведите на экран списки "gen_1" и "gen_2"




````python
# Код начинается здесь ...
gen_1 = [input() for _ in range(5)]
gen_2 = [int(i) for i in gen_1 if i.isnumeric()]

print(gen_1)
print(gen_2)
````

# Задание №3
У Вас имеется некий массив данных:
```
a = [2,-2,4,-4,7,5]
```
Вам необходимо с помощью генератора возвести в квадрат каждое значение.<br>
Результат запишите в переменную "b" и выведите на экран





````python
# Код начинается здесь ...
a = [2,-2,4,-4,7,5]

b1 = [i**2 for i in a]
b2 = [pow(i,2) for i in a]
b3 = [i*i for i in a]

print(b1)
print(b2)
print(b3)
````
