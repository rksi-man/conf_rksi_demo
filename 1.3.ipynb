–í—ã–≤–æ–¥–∏—Ç –ø–æ–ª—É—á–∏–≤—à–∏–π—Å—è —Å–ø–∏—Å–æ–∫ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ –¥–µ—Å—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Ä—è–¥.
–î–∞–ª–µ–µ —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ sort() –∏ —Å–Ω–æ–≤–∞ –≤—ã–≤–æ–¥–∏—Ç –µ–≥–æ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ –¥–µ—Å—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ, —Ç–∞–∫–∂–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç –ø–æ–≤—Ç–æ—Ä—è—é—â–∏—Ö—Å—è –∑–Ω–∞—á–µ–Ω–∏–π. –î–ª—è –≤—ã–≤–æ–¥–∞ —Å–ø–∏—Å–∫–∞ –Ω–∞–ø–∏—à–∏—Ç–µ –æ—Ç–¥–µ–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é, –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –æ–Ω–∞ –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Å–ø–∏—Å–æ–∫.





````python
# –ö–æ–¥ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –∑–¥–µ—Å—å ...
import random
````

````python
'''
–°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å —Å–ª—É—á–∞–π–Ω—ã–º–∏
–∑–Ω–∞—á–µ–Ω–∏—è–º–∏. –†–∞–∑–º–µ—Ä 100
–î–∏–∞–ø–∞–∑–æ–Ω —Å–ª. —á–∏—Å–µ–ª 1-200
'''

rand_nums = []

for i in range(100):
  num = random.randint(1,200)
  rand_nums.append(num)
````

````python
def print_list(list_r):
  '''
  –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ —Å–ø–∏—Å–∫–∞ –Ω–∞ —ç–∫—Ä–∞–Ω
  –ø–æ 10 —ç–ª. –≤ —Å—Ç—Ä–æ–∫–µ
  '''
  for i in range(0,len(list_r),10):
    print(list_r[i:i+10])
````

````python
print_list(rand_nums)
````

````python
rand_nums_unique = list(set(rand_nums))
````

````python
rand_nums_unique.sort()
````

````python
print_list(rand_nums_unique)
````
