# Задание №7
Напишите программу, которая запрашивала бы у пользователя:
- его имя (например, "Как тебя зовут?")
- возраст ("Сколько тебе лет?")
- место жительства(город) ("Где ты живешь?") <br>
>После этого выводила бы три строки:
```
Привет ИМЯ!
Тебе ВОЗРАСТ
Ты живешь в ГОРОД
```

Вместо *ИМЯ*, *ВОЗРАСТ*, *ГОРОД* должны быть данные, введенные
пользователем.





````python
# Код начинается здесь ...
var_name = input('Как тебя зовут?\n')
var_age = input('Сколько тебе лет?\n')
var_place = input('Где ты живешь?\n')

print('Привет', var_name + '!')
print('Тебе', var_age)
print('Ты живешь в', var_place)
````
